```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as snc
```

```python
is_101_T = pd.read_csv("data/is_101_T(csv).csv")
```

```python
is_101_T.head()
```

```python
is_101_T.dtypes
```

```python
is_101_T['Date'] = is_101_T['Date'].astype(str)
```

```python
is_101_T['Date']=is_101_T['Date'].str.replace('-','/')
is_101_T.head()
```

```python
from datetime import datetime
Date=[]
for i in is_101_T['Date']:
    Date.append(datetime.strptime(i, '%Y/%m/%d'))
Date
```

```python
date=pd.Series(Date)
date
```

```python
is_101_T['Date']=pd.to_numeric(date)
is_101_T
```

```python
is_101_T['Date']=is_101_T['Date'].astype(float)
is_101_T.dtypes
```

```python
from datetime import datetime
Time = []
for i in is_101_T['Time']:
    Time.append(datetime.strptime(i, '%H.%M.%S'))
is_101_T.dtypes
```

```python
is_101_T
```

```python
time=pd.Series(Time)
is_101_T['Time']=pd.to_numeric(time)
```

```python
is_101_T['Time']=is_101_T['Time'].astype(float)
is_101_T.dtypes
```

```python
is_101_T.tail()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
num = is_101_T.keys()
scaler = MinMaxScaler()
scaler.fit(is_101_T[num])
is_101_T[num] = scaler.transform(is_101_T[num])
```

```python
snc.pairplot(is_101_T[["Date","Time","SO2","NO2","O3","CO","PM10","PM2.5"]], diag_kind = 'auto')
```

```python
features=is_101_T
target=is_101_T['T']
```

```python
is_101_T2 = is_101_T.corr('pearson')
is_101_T2
```

```python
abs(is_101_T2['T']).sort_values(ascending=False)
```

```python
is_101_T['T'].value_counts().plot(kind='bar', color='lightblue',figsize=(50,10))
```

```python
# 상관관계가 낮은 값을 갖는 행을 제거

features=features.drop('Date',axis=1)
features=features.drop('Time',axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target)
y_test.tail()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression(normalize=True)
regressor.fit(X_train, y_train)
```

```python
print("Predicted values:", regressor.predict(X_test))
y_pred = regressor.predict(X_test)
y_pred.shape
```

```python
print("R^2 score for linear regression: ", regressor.score(X_test, y_test))
```

```python
plt.scatter(y_test, y_pred, color='black')
plt.show()
```
