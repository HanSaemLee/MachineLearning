```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as snc
```

```python
air_quality = pd.read_csv("data/AirQualityUCI.csv", sep=";", decimal=",")
air_quality
```

```python
air_quality.dropna(axis=0, how= 'all', inplace=True)
air_quality.dropna(axis=1, inplace=True)
air_quality
```

```python
air_quality.replace(to_replace= -200, value = np.NaN, inplace= True)
air_quality
```

```python
air_quality.fillna(air_quality.mean(), inplace= True)
air_quality
```

```python
snc.heatmap(air_quality.isnull(),yticklabels=False, cbar=False, cmap='viridis')
```

```python
# loc는 인간이 읽을 수 있는 값으로 데이터에 접근
# 레이블 이름이 Date인 열에 접근
air_quality.loc[:,'Date']=air_quality['Date']
air_quality
```

```python
air_quality.dtypes
```

```python
air_quality['Date']
```

```python
# 날짜와 시간 변환
from datetime import datetime
Date= []
for i in air_quality['Date']:
    Date.append(datetime.strptime(i, '%d/%m/%Y'))   # strptime은 문자열을 날짜/시간으로 변환
Date
```

```python
date=pd.Series(Date)
date
```

```python
air_quality['Date']=pd.to_numeric(date)
air_quality['Date']
```
