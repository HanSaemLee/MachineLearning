```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as snc
```

```python
air_quality = pd.read_csv("data/Measurement_summary.csv")
```

```python
air_quality.head()
```

```python
air_quality.dropna(axis=0, how= 'all', inplace=True)
air_quality.dropna(axis=1, inplace=True)
air_quality.head()
```

```python
air_quality.replace(to_replace= -200, value = np.NaN, inplace= True)
air_quality.head()
```

```python
snc.heatmap(air_quality.isnull(),yticklabels=False, cbar=False, cmap='viridis')
```

```python
air_quality.loc[:,'Measurement date']= air_quality['Measurement date']
air_quality
```

```python
air_quality.dtypes
```

```python
air_quality['Measurement date']
```

```python
# type을 object --> datetime 변경
air_quality['Date_datetime']=pd.to_datetime(air_quality['Measurement date'])
```

```python
# 끝에 Date_datetime 열이 생성됨
air_quality.head()
```

```python
# date, time, year, month, day 이렇게 뽑을 수 있음
air_quality['Date']=air_quality['Date_datetime'].dt.date
air_quality['Time']=air_quality['Date_datetime'].dt.time
```

```python
air_quality.head()
```

```python
air_quality.dtypes
```

```python
# 열 삭제
air_quality = air_quality.drop(['Measurement date', 'Date_datetime', 'Address', 'Latitude', 'Longitude'], axis=1)
air_quality.head()
```

```python
# 열 순서 변경
air_quality = air_quality[['Station code','Date','Time','SO2','NO2','O3','CO','PM10','PM2.5']]
air_quality
```

```python
air_quality.dtypes
```

```python
# Time열을 문자열로 바꾸어야 .str.replace가 된다
air_quality['Time'] = air_quality['Time'].astype(str)
```

```python
air_quality['Time']=air_quality['Time'].str.replace(':', '.')
air_quality.head()
```

```python
air_quality['Date'] = air_quality['Date'].astype(str)
```

```python
air_quality['Date']=air_quality['Date'].str.replace('-', '/')
air_quality.head()
```
